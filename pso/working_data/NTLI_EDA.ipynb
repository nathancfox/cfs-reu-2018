```python
ntli = ntli.rename(columns={'Compound Name': 'Compound',
                            'Stratified screen category (1=hit, 0=non-hit)': 'Hit'})
ntli.head()
```

Checking to see if there are any NaN values. [source](https://stackoverflow.com/questions/29530232/how-to-check-if-any-value-is-nan-in-a-pandas-dataframe)

```python
ntli.isnull().sum().sum()
```

```python
hit_ntli = ntli[ntli['Hit'] == 1]
miss_ntli = ntli[ntli['Hit'] == 0]
```

```python
hit_ntli.head()
```

```python
miss_ntli.head()
```

```python
del hit_ntli['Hit']
del miss_ntli['Hit']
```

```python
miss_ntli.reset_index(drop=True, inplace=True)
miss_ntli.head()
```

```python
hit_ntli.set_index('Compound', inplace=True)
```

```python
miss_ntli.set_index('Compound', inplace=True)
```

```python
zero_responses = []
for i in range(hit_ntli.shape[0]):
    zero_responses.append(hit_ntli.iloc[i].value_counts().loc[0.0])
hit_ntli['Zero_Responses'] = zero_responses
hit_ntli.head()
```

```python
zero_responses = []
for i in range(miss_ntli.shape[0]):
    zero_responses.append(miss_ntli.iloc[i].value_counts().loc[0.0])
miss_ntli['Zero_Responses'] = zero_responses
miss_ntli.head()
```

```python
fig1 = plt.figure(figsize=(20, 10))
hit_ax1 = fig1.add_subplot(121)
hit_ax1.hist(hit_ntli['Zero_Responses'], bins=[i*20 for i in range(9)])
hit_ax1.set_title('Hits')
hit_ax1.set_ylabel('Frequency')
hit_ax1.set_xlabel('Number of 0 Response Kinases (out of 190)')
miss_ax1 = fig1.add_subplot(122, sharey=hit_ax1)
miss_ax1.hist(miss_ntli['Zero_Responses'], bins=[i*20 for i in range(9)])
miss_ax1.set_title('Non-Hits')
miss_ax1.set_ylabel('Frequency')
miss_ax1.set_xlabel('Number of 0 Response Kinases (out of 190)')
plt.show()
```

```python
pd.concat([hit_ntli['Zero_Responses'], miss_ntli['Zero_Responses']], axis=1).describe()
```

```python
hit_ntli['Zero_Responses_Percent'] = hit_ntli['Zero_Responses'] / 190
miss_ntli['Zero_Responses_Percent'] = miss_ntli['Zero_Responses'] / 190
```

```python
fig2 = plt.figure(figsize=(20, 10))
hit_ax2 = fig2.add_subplot(121)
hit_ax2.hist(hit_ntli['Zero_Responses_Percent'], bins=[i*0.1 for i in range(11)])
hit_ax2.set_title('Hits')
hit_ax2.set_ylabel('Frequency')
hit_ax2.set_xlabel('Percent of 0 Response Kinases (out of 190)')
miss_ax2 = fig2.add_subplot(122, sharey=hit_ax2)
miss_ax2.hist(miss_ntli['Zero_Responses_Percent'], bins=[i*0.1 for i in range(11)])
miss_ax2.set_title('Non-Hits')
miss_ax2.set_ylabel('Frequency')
miss_ax2.set_xlabel('Percent of 0 Response Kinases (out of 190)')
plt.show()
```
