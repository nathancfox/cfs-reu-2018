```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('var_by_time.csv', index_col=0)
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(20, 20))
ax1 = fig.add_subplot(211)
ax1.plot(df.index, df['num_features'], 'ro')
ax1.legend()
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Number of Features in abinary')
ax2 = fig.add_subplot(212)
ax2.plot(df.index, df['g_fitness'], 'bx')
ax2.plot(df.index, df['a_fitness'], 'gv')
ax2.plot(df.index, df['a_score'], 'ro')
ax2.legend()
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Fitness Function')
plt.show()
```
