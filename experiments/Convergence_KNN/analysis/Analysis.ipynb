```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

plt.style.use('ggplot')
```

```python
vbt00 = pd.read_csv('../00_report_spread/var_by_time.csv', index_col=0)
print(vbt00.shape)
vbt00
```

```python
fig, ax = plt.subplots()
ax.plot(vbt00.index, vbt00['spread_sum'], '-o')
ax.set_title('Sum of Particle Distances From Swarm Center')
ax.set_xlabel('Time')
ax.set_ylabel('Sum of Distances')
axin = zoomed_inset_axes(ax, 3, loc=10)
axin.plot(vbt00.index, vbt00['spread_sum'], '-o')
axin.set_xlim(0, 15)
axin.set_ylim(0, 1000)
axin.set_fc('w')
axin.grid(color='#A0A0A0')
mark_inset(ax, axin, loc1=2, loc2=3, fc="none", ec="0.3")
#plt.savefig('../../lab_notebooks/cfs_notebook_files/Swarm_Convergence_PLOT_Sum_Distance.svg')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(vbt00.index, vbt00['spread_mean'], '-o')
ax.set_title('Mean Particle Distance From Swarm Center')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Distance')
axin = zoomed_inset_axes(ax, 3, loc=10)
axin.plot(vbt00.index, vbt00['spread_mean'], '-o')
axin.set_xlim(0, 15)
axin.set_ylim(0, 10)
axin.set_fc('w')
axin.grid(color='#A0A0A0')
mark_inset(ax, axin, loc1=2, loc2=3, fc="none", ec="0.3")
plt.savefig('../../lab_notebooks/cfs_notebook_files/Swarm_Convergence_PLOT_Mean_Distance.svg')
plt.show()
```
