```python
timesteps = df00.shape[0]
for i in range(timesteps):
    plt.plot(df00.index[:timesteps], df00.iloc[:timesteps, :])
plt.gcf().set_size_inches(15, 10)
plt.ylim(0.0, 1.0)
plt.ylabel('Fraction of Velocity that is "Extreme"')
plt.xlabel('Time')
plt.title('Extremeness of Velocity - Full Swarm')
plt.savefig('../../lab_notebooks/cfs_notebook_files/Mean_Particle_Velocity_Sequel_PLOT_Full.png')
plt.show()
```

```python
var_by_time = pd.read_csv('../00_report_velocities/var_by_time.csv', index_col = 0)
print(var_by_time.shape)
var_by_time.head()
```

```python
plt.plot(var_by_time.index, var_by_time['num_features'], '-o')
plt.xlabel('Time')
plt.ylabel('Number of Features in abest')
plt.title('Number of Features')
plt.gcf().set_size_inches(15, 10)
```

```python
gfig = plt.figure(figsize=(15, 15))
gax = gfig.add_subplot(211)
gax.plot(var_by_time.index, var_by_time['g_fitness'], '-o')
gax.plot(var_by_time.index, var_by_time['g_score'], '-x')
aax = gfig.add_subplot(212)
aax.plot(var_by_time.index, var_by_time['a_fitness'], '-o')
aax.plot(var_by_time.index, var_by_time['a_score'], '-x')
aax.set_xlabel('Time')
gax.set_ylabel('Fitness or Accuracy')
aax.set_ylabel('Fitness or Accuracy')
gax.set_title('gbest Statistics')
aax.set_title('abest Statistics')
aax.set_ylim(0.67, 0.85)
gax.set_ylim(0.67, 0.85)
aax.legend()
gax.legend()
gfig.savefig('../../lab_notebooks/cfs_notebook_files/Mean_Particle_Velocity_Sequel_PLOT_Fitness_Score_AG.svg')
```
