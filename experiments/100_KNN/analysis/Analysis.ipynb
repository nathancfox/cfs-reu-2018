```python
kin = []
with open('reported_kinases.csv', 'r') as f:
    for line in f:
        kin.append(line.strip().split(','))
```

```python
kin
```

```python
k_cnts = {}
for l in kin:
    for k in l:
        if k in k_cnts:
            k_cnts[k] += 1
        else:
            k_cnts[k] = 1
```

```python
k_cnts
```

```python
kin_cnt_ser = pd.Series(list(k_cnts.values()), index=k_cnts.keys())
```

```python
kin_cnt_ser.sort_values(ascending=False, inplace=True)
```

```python
kin_cnt_ser
```

```python
groups = []
with open('linked_groups.csv', 'r') as f:
    for line in f:
        groups.append(line.strip().split(','))
for i in range(len(groups)):
    groups[i] = set(groups[i])
```

```python
groups
```

```python
sum = 0
for g in groups:
    sum += len(g)
print(sum)
print(190-sum)
```

```python
new_col = [-2 for i in range(len(kin_cnt_ser.index))]
for i in range(len(kin_cnt_ser.index)):
    for j in range(len(groups)):
        if kin_cnt_ser.index[i] in groups[j]:
            new_col[i] = j
            continue
    if new_col[i] == -2:
        new_col[i] = -1
```

```python
-2 in new_col
```

```python
kin_cnt_df = pd.DataFrame({'Count': kin_cnt_ser, 'Group': new_col})
```

```python
in_groups = np.zeros(1600).reshape((100, 16)).astype(bool)
```

```python
for i in range(len(kin)):
    for j in range(len(kin[i])):
        for k in range(len(groups)):
            if kin[i][j] in groups[k]:
                in_groups[i, k] = True
                break
        else:
            in_groups[i, -1] = True
```

```python
group_representation = np.zeros(16)
for i in range(in_groups.shape[0]):
    for j in range(in_groups.shape[1]):
        if in_groups[i, j]:
            group_representation[j] += 1
```

```python
group_representation
```

```python
group_frequencies = group_representation / np.array([len(g) for g in groups] + [144])
```

```python
group_frequencies
```

```python
plt.style.use('ggplot')
labels = [str(i+1) for i in range(15)] + ['N/A']
fig, ax = plt.subplots()
fig.set_size_inches((12, 8.25))
ax.bar([i+1 for i in range(16)], group_frequencies, tick_label=labels, align='center', color='#2f8cb5')
ax.set_xlabel('Group Number', size=14)
ax.set_xlim((0, 17))
ax.set_ylabel('Frequency of Inclusion (Normalized to Group Size)', size=14)
ax.set_title('Frequency of Inclusion of At Least 1 Member of Each Group',
             size='20')
ax.tick_params(axis='both', pad=8, labelsize=12)
fig.set_facecolor('#ffffff')
plt.savefig('figures/Frequency_of_Inclusion_Normalized.png', dpi=300, format='png',
            bbox_inches='tight')
plt.show()
```

```python
scores = pd.read_csv('numerical_data.csv')
```

```python
scores.head()
```

```python
scores.drop(['Unnamed: 0', 'g_fitness', 'g_accuracy', 'g_sensitivity', 'g_specificity'], axis=1, inplace=True)
```

```python
scores.head()
```

```python
scores.describe()
```

```python
acc = [75.1111, 79.3846, 85.564, 82.3111, 85.9777, 83.3386, 81.9413,79.7157, 82.0088, 82.2484]
sen = [57, 71.9524, 86.6667, 75.8095, 84.2857, 78.6667, 72.381, 78.0952, 72.6667, 66.9524]
spe = [80.7143, 82.8849, 84.4286, 84.3636, 86.619, 85.119, 86.9381, 81.7932, 84.7179, 86.2098]
```

```python
t, p = stats.ttest_ind(acc, scores['a_accuracy'])
print('Accuracy\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
print()
t, p = stats.ttest_ind(sen, scores['a_sensitivity'])
print('Sensitivity\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
print()
t, p = stats.ttest_ind(spe, scores['a_specificity'])
print('Specificity\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
```

```python
control_scores = pd.read_csv('control_scores.csv')
control_scores.head()
```

```python
fig = plt.figure(figsize=(10, 10))
ax_acc = fig.add_subplot(311)
ax_acc.set_title('Scores')
ax_acc.errorbar([1, 2], [control_scores['accuracy'].mean(), scores['a_accuracy'].mean()], yerr = [control_scores['accuracy'].std(), scores['a_accuracy'].std()], linestyle='None', marker='o', markerfacecolor='r', ms=10, ecolor='k', capsize=10)
ax_acc.set_xlim((0.5, 2.5))
ax_acc.set_xticks([1, 2])
ax_acc.set_xticklabels(['', ''])
ax_acc.set_ylabel('Accuracy')
ax_sen = fig.add_subplot(312, sharex=ax_acc)
ax_sen.errorbar([1, 2], [control_scores['sensitivity'].mean(), scores['a_sensitivity'].mean()], yerr = [control_scores['sensitivity'].std(), scores['a_sensitivity'].std()], linestyle='None', marker='o', markerfacecolor='g', ms=10, ecolor='k', capsize=10)
ax_sen.set_xlim((0.5, 2.5))
ax_sen.set_xticks([1, 2])
ax_sen.set_xticklabels(['', ''])
ax_sen.set_ylabel('Sensitivity')
ax_spe = fig.add_subplot(313, sharex=ax_acc)
ax_spe.errorbar([1, 2], [control_scores['specificity'].mean(), scores['a_specificity'].mean()], yerr = [control_scores['specificity'].std(), scores['a_specificity'].std()], linestyle='None', marker='o', markerfacecolor='b', ms=10, ecolor='k', capsize=10)
ax_spe.set_xlim((0.5, 2.5))
ax_spe.set_xticks([1, 2])
ax_spe.set_ylabel('Specificity')
ax_spe.set_xticklabels(['Control', 'Experimental'])
plt.savefig('figures/scores.png', dpi=300, format='png')
plt.show()
```

```python
t, p = stats.ttest_ind(control_scores['accuracy'], scores['a_accuracy'])
print('Accuracy\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
print()
t, p = stats.ttest_ind(control_scores['sensitivity'], scores['a_sensitivity'])
print('Sensitivity\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
print()
t, p = stats.ttest_ind(control_scores['specificity'], scores['a_specificity'])
print('Specificity\n'+'-'*40)
print('    t: {}'.format(t))
print('    p: {}'.format(p))
```
