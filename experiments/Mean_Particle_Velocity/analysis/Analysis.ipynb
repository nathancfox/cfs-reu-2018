```python
steps = df00.shape[0]
for i in range(df00.shape[1]):
    plt.plot([i for i in range(steps)], df00.iloc[:steps, :], '-x')
plt.gcf().set_size_inches(15, 10)
plt.title('Mean Velocity for 100 Particles')
plt.xlabel('Time')
plt.ylabel('Mean of Values in Velocity Vector')
plt.savefig('../../lab_notebooks/cfs_notebook_files/Mean_Particle_Velocity_PLOT_Full.svg')
#plt.show()
```
